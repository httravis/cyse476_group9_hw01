Part 1 - Words to Patterns:

1.	For any given string of characters, return the case-independent pattern (we'll discuss in class). 
For example, the word "apple" may be represented as the pattern "abbcd"
The word "zoozoo" may be "abbabb"
The word "Banana" and "BAnana" both have the pattern "abcbcb" where "BANANAS" has the pattern "abcbcbd"

Requirement: The program should be able to take a string from console (typed word) or read words from a file


```python
from collections import OrderedDict # This could be very helpful in HW 01

def c_az(key):
    c = ''.join(OrderedDict.fromkeys(key).keys())
    return c

az = 'abcdefghijklmnopqrstuvwxyz'
azlen = len(az)
print(az)
print(azlen)

word = "Banana"
word = word.replace(' ','').lower()
print(word)

word_index = c_az(word)
print(word_index)

w_pattern = ''
for c in range(len(word)):
    wchar = word[c]
    windex = word_index.find(word[c])
    print(c,'\t',wchar,'\t',windex,'\t',az[windex]) 
    w_pattern += az[windex] 

print(word)
print(w_pattern)
```
